```python
import numpy as np
import scipy.stats as stats

Football_Match = [377,356.5,466,787.5,710,535.5,710.5,553.5,383.5,393.5,389.5,467.5,425.5,597,556,464,719,507.5,362,338,300.5,305,615,589,533.5,578.5599,530.5,476.5,241.5,337.5,360.5,491.5,485.5,677.5,516.5,558,571,216.5,144.5,208.5,312.5,361.5,561.5,547,368.5,506,476.5,267,137.5,279,319.5,435,669,481,407.5,560,524.5,383.5,238.5,402,424.5,661.5,378,608,565.5,617.5,457,417.5,236.5,187.5,294.5,277.5,354,441.5,532.5,402.5,320,376,204,177,226.5,482,354,523,479.5,446.5,467,289,254.5,554.5]
Non_Football_Match = [330.2, 343.8, 510.4, 532, 710.2, 657, 607.6, 507, 394.8, 299.4, 367.2, 357, 372.2, 472.6, 571.2, 541.8, 555.2, 483, 408.2, 238, 324, 304.8, 452.8, 597.2, 529, 527, 539.8, 542.8, 351.8, 253.8, 223, 242, 491.4, 524.8, 504.6, 476.2, 544, 430.2, 275.8, 166.6, 162.8, 217.6, 415.2, 489.6, 536.4, 480.8, 467.4, 454, 291.4, 121.2, 260.2, 296.4, 496, 510, 534, 452, 480.6, 416.2, 363.2, 201.8, 387.4, 404.4, 616.2, 394.2, 604.4, 532.8, 539.6, 423.4, 369.6, 229, 195.4, 270.2, 275.6, 390.2, 403.6, 453.2, 429.8, 364.4, 290.8, 232.4, 181, 237.6, 386.8, 436.6, 400.8, 394.6, 512.6, 342.4, 276.6, 256.6]

data = [Football_Match, Non_Football_Match]

grand_mean = np.mean(Football_Match + Non_Football_Match)

SSA = len(Football_Match) * (np.mean(Football_Match) - grand_mean) ** 2 + len(Non_Football_Match) * (np.mean(Non_Football_Match) - grand_mean) ** 2

SST = sum((x - grand_mean) ** 2 for x in Football_Match + Non_Football_Match)

SSE = SST - SSA

df_between = len(data) - 1 
df_within = len(Football_Match) + len(Non_Football_Match) - 2  
df_total = len(Football_Match) + len(Non_Football_Match) - 1 

MSA = SSA / df_between  
MSE = SSE / df_within 

F_statistic = MSA / MSE

F, p_value = stats.f_oneway(Football_Match, Non_Football_Match)

correlation = np.corrcoef(Football_Match, Non_Football_Match)[0, 1]

print("Grand Mean:", round(grand_mean, 4))
print("SSA (Sum of Squares Between Groups):", round(SSA, 4))
print("SST (Total Sum of Squares):", round(SST, 4))
print("SSE (Sum of Squares Error):", round(SSE, 4))
print("Degrees of Freedom (Between, Within, Total):", df_between, df_within, df_total)
print("Mean Square Between (MSA):", round(MSA, 4))
print("Mean Square Error (MSE):", round(MSE, 4))
print("F-statistic:", round(F_statistic, 4))
print("F-statistic from scipy ANOVA:", round(F, 4))
print("P-value from ANOVA:", round(p_value, 4))
print("Correlation Coefficient (C):", round(correlation, 4))

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. The means are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis. The means are not significantly different.")
```
